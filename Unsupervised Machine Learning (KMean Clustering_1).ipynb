```python
import seaborn as sns 
```

```python
dataset = sns.load_dataset("iris")
dataset.columns
```

```python
new_data = dataset.loc[:,["sepal_length","petal_length"]]
```

```python
new_data
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=new_data)
```

```python
from sklearn.cluster import KMeans
```

```python
er = []
no_cl = [i for i in range(2,20)]
```

```python
for i in range(2,20):
    kr = KMeans(n_clusters=i)
    kr.fit(new_data)
    er.append(kr.inertia_)
```

```python
new_data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
plt.plot(no_cl,er,marker="o",c="r")
plt.grid()
plt.xticks([i for i in range(1,21)])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
ka = KMeans(n_clusters=3)
```

```python
new_data["output"]=ka.fit_predict(new_data)
```

```python
new_data
```

```python
sns.scatterplot(x="sepal_length",y="petal_length",data=new_data,hue="output")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dataset["species_new"] = le.fit_transform(dataset["species"])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(dataset["species_new"],new_data["output"])*100
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("pred")
sns.scatterplot(x="sepal_length",y="petal_length",data=new_data,hue="output")
plt.subplot(1,2,2)
sns.scatterplot(x="sepal_length",y="petal_length",data=dataset,hue="species_new")
plt.title("org")
```
